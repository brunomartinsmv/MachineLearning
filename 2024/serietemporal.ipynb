```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
from datetime import datetime
```

```python
stock2 = pd.read_csv("C:/Users/Dell/Documents/all-stocks-2006.csv", parse_dates=['Date'])
```

```python
stock2.head(30)
```

```python
stock2['Date'] = pd.to_datetime(stock2['Date'])
```

```python
google = stock2[stock2['Name'] == 'GOOGL']
google.head(30)
```

```python
mmm = stock2[stock2['Name'] == 'MMM']
mmm.head(30)
```

```python
google = google.copy()
google['Ticks'] = range(0, len(google.index.values))
google.head(30)
```

```python
mmm = mmm.copy()
mmm['Ticks'] = range(0, len(mmm.index.values))
mmm.head(30)
```

```python
fig = plt.figure(1)

ax1 = fig.add_subplot(121)
ax1.set_xlabel('Ticks')
ax1.set_ylabel('Preo')
ax1.set_title('Stock Google')
ax1.plot('Ticks', 'Close', data=google)

ax2 = fig.add_subplot(122)
ax2.set_xlabel('Ticks')
ax2.set_ylabel('Preo')
ax2.set_title('Stock 3M')
ax2.plot('Ticks', 'Close', data=mmm, color='red')
plt.show()


```

```python
google2=google.copy()
google2['Movel'] = google2['Close'].rolling(window=10).mean()
google2.head(30)
```

```python
filt = google2[(google2['Ticks'] >= 900) & (google2['Ticks'] <= 3000)]
filt.head(30)

filt2 = google[(google['Ticks'] >= 900) & (google['Ticks'] <= 1100)]    
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(20,10))
axes[0].plot('Ticks','Open',data = google)
axes[0].set_title('Original')
axes[1].plot('Ticks','Open',data = filt2)
axes[1].set_title('Mdia Mvel')
```
